```python
import os
import pandas as pd
import snowflake.connector
```

```python
kpi_query = """
SELECT
  COUNT(*) AS total_customers,
  SUM(IFF(Churn='Yes',1,0)) AS churned_customers,
  ROUND(100.0 * AVG(IFF(Churn='Yes',1,0)), 2) AS churn_rate_percent,
  ROUND(AVG(MonthlyCharges), 2) AS avg_monthly_charges,
  ROUND(AVG(IFF(Churn='Yes', MonthlyCharges, NULL)), 2) AS avg_monthly_charges_churned
FROM SNOWFLAKE_LEARNING_DB.PUBLIC.TELCO;
"""
kpi = pd.read_sql(kpi_query, conn)
kpi
```

```python
df_contract = pd.read_sql(""" 
 select contract,
 round(100 * avg(iff(churn='churned',1,0)),2)as churn_rate_percent
 FROM SNOWFLAKE_LEARNING_DB.PUBLIC.TELCO
 group by contract
 order by churn_rate_percent desc
 """, conn)
df_internet = pd.read_sql("""
SELECT
  InternetService,
  ROUND(100.0 * AVG(IFF(Churn='Churned',1,0)), 2) AS churn_rate_percent,
  COUNT(*) AS customers
FROM SNOWFLAKE_LEARNING_DB.PUBLIC.TELCO
GROUP BY InternetService
ORDER BY churn_rate_percent DESC;
""", conn)

df_tenure = pd.read_sql("""
SELECT
  CASE
    WHEN tenure < 6 THEN '0-5'
    WHEN tenure < 12 THEN '6-11'
    WHEN tenure < 24 THEN '12-23'
    WHEN tenure < 48 THEN '24-47'
    ELSE '48+'
  END AS tenure_bucket,
  ROUND(100.0 * AVG(IFF(Churn='Churned',1,0)), 2) AS churn_rate_percent,
  COUNT(*) AS customers
FROM SNOWFLAKE_LEARNING_DB.PUBLIC.TELCO
GROUP BY tenure_bucket
ORDER BY tenure_bucket;
""", conn)

df_charge_dist = pd.read_sql("""
SELECT
  Churn,
  MonthlyCharges
FROM SNOWFLAKE_LEARNING_DB.PUBLIC.TELCO
WHERE MonthlyCharges IS NOT NULL;
""", conn)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(df_contract["CONTRACT"], df_contract["CHURN_RATE_PERCENT"])
plt.xlabel("Contract")
plt.ylabel("Churn rate (%)")
plt.title("Churn Rate by Contract")
plt.xticks(rotation=20, ha="right")
plt.tight_layout()
plt.show()
```

```python
#INSIGHT
Month to month shows highest churn rate, indicating customers without long commitment are likely to leave.
```

```python
plt.figure()
plt.bar(df_internet["INTERNETSERVICE"], df_internet["CHURN_RATE_PERCENT"])
plt.xlabel("Internet Service")
plt.ylabel("Churn rate (%)")
plt.title("Churn Rate by Internet Service")
plt.xticks(rotation=20, ha="right")
plt.tight_layout()
plt.show()
```

```python
#INSIGHT
Fiber optic customers exhibit the highest churn rates among
internet service categories. This may indicate dissatisfaction
with cost-performance balance or service expectations.

In contrast, DSL and non-internet customers show lower churn,
suggesting stability among lower-tier or less service-dependent users.

Targeted retention strategies such as pricing optimization or
service quality improvements could reduce churn in high-value segments.
```

```python
plt.figure()
plt.plot(df_tenure["TENURE_BUCKET"], df_tenure["CHURN_RATE_PERCENT"], marker="o")
plt.xlabel("Tenure bucket (months)")
plt.ylabel("Churn rate (%)")
plt.title("Churn Rate by Tenure Bucket")
plt.tight_layout()
plt.show()
```

```python
#INSIGHT
Analysis of churn across tenure buckets shows a clear inverse
relationship between customer lifespan and churn probability.
New customers (0–12 months) exhibit the highest churn rates,
while long-term customers demonstrate strong retention.

This indicates that early-stage engagement is critical.
Improving onboarding experience, customer support, and
initial value delivery could significantly reduce churn risk.
```

```python
plt.figure()
yes = df_charge_dist[df_charge_dist["CHURN"] == "Churned"]["MONTHLYCHARGES"]
no  = df_charge_dist[df_charge_dist["CHURN"] == "Stayed"]["MONTHLYCHARGES"]

plt.hist(no, bins=30, alpha=0.6, label="Stayed")
plt.hist(yes, bins=30, alpha=0.6, label="Churned")

plt.xlabel("Monthly Charges")
plt.ylabel("Customers")
plt.title("Monthly Charges Distribution (Churn vs Stayed)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#INSIGHT
Analysis of monthly billing patterns indicates that churned
customers generally fall within higher charge ranges compared
to retained customers. This highlights potential price
sensitivity among premium-paying users.

Retention strategies such as personalized offers, loyalty
discounts, or enhanced service value could help preserve
high-revenue customers and reduce overall churn impact.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

heat_df = pd.read_sql("""
SELECT
  Contract,
  InternetService,
  AVG(IFF(Churn='Churned',1,0)) AS churn_rate
FROM SNOWFLAKE_LEARNING_DB.PUBLIC.TELCO
GROUP BY Contract, InternetService
""", conn)

pivot = heat_df.pivot(
    index="CONTRACT",
    columns="INTERNETSERVICE",
    values="CHURN_RATE"
)

plt.figure()
sns.heatmap(pivot, annot=True)
plt.title("Churn Rate Heatmap")
plt.show()
```

```python
#INSIGHT
Interaction analysis between contract type and internet service
reveals concentrated churn risk among fiber optic customers on
short-term contracts. This suggests that high-cost or
high-expectation services without long-term commitment lead to
increased switching behaviour.

Retention efforts should prioritize this segment through
pricing incentives, service quality monitoring, or contract
conversion campaigns.

Conversely, customers on multi-year agreements display strong
stability regardless of service type, reinforcing the value of
long-term engagement strategies.
```

```python
sns.boxplot(
    data=df_charge_dist,
    x="CHURN",
    y="MONTHLYCHARGES"
)
plt.title("Monthly Charges vs Churn")
plt.show()
```

```python
#INSIGHT
Distribution analysis shows churned users skew toward higher
monthly billing levels, with a broader spread of charges.
Retained customers cluster around lower charge ranges,
indicating price sensitivity within the customer base.

Protecting high-revenue customers through loyalty incentives,
bundled value offerings, or personalized pricing may reduce
revenue loss while stabilizing retention.
```

```python
df_ml = pd.read_sql("""
SELECT *
FROM SNOWFLAKE_LEARNING_DB.PUBLIC.TELCO
""", conn)

df_ml.head()
```

```python
df_ml["CHURN_FLAG"] = df_ml["CHURN"].map({
    "Stayed": 0,
    "Churned": 1
})
```

```python
features = [
    "TENURE",
    "MONTHLYCHARGES"
]

X = df_ml[features]
y = df_ml["CHURN_FLAG"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

preds = model.predict(X_test)

accuracy_score(y_test, preds)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_estimator(
    model,
    X_test,
    y_test
)
```

```python

importance = pd.Series(
    model.coef_[0],
    index=features
)

importance.sort_values().plot.barh()
```

```python
features = [
    "TENURE",
    "MONTHLYCHARGES",
    "CONTRACT",
    "INTERNETSERVICE",
    "PAYMENTMETHOD"
]

X = df_ml[features]
y = df_ml["CHURN_FLAG"]
```

```python
X = pd.get_dummies(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42
)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=8,
    random_state=42
)

rf.fit(X_train, y_train)
```

```python

preds = rf.predict(X_test)
accuracy_score(y_test, preds)
```

```python
ConfusionMatrixDisplay.from_estimator(
    rf,
    X_test,
    y_test
)
```

```python
importance = pd.Series(
    rf.feature_importances_,
    index=X.columns
).sort_values()

importance.tail(10).plot.barh()
```

```python
from sklearn.metrics import (
    classification_report,
    roc_auc_score,
    roc_curve
)
import matplotlib.pyplot as plt
```

```python
print(classification_report(y_test, preds))
```

```python
auc = roc_auc_score(y_test, preds)
print("AUC Score:", auc)
```

```python
fpr, tpr, _ = roc_curve(y_test, preds)

plt.figure()
plt.plot(fpr, tpr)
plt.plot([0,1],[0,1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.show()
```
